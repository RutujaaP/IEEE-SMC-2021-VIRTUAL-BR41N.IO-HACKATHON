AFTER HYPERPARAMETER TUNING

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform as sp_randFloat
from scipy.stats import randint as sp_randInt
```

```python
model=CatBoostClassifier()
parameters={ 'depth'         : sp_randInt(4,10),
              'learning_rate':sp_randFloat(),
              'iterations'   :sp_randInt(10,100)
              # 'colsample_bylevel': [0.3, 0.8, 0.1]
           }
randm=RandomizedSearchCV(estimator=model,param_distributions=parameters, cv=2, n_iter=10, n_jobs=-1)
randm.fit(X_train,Y_train)
```

```python
print(randm.best_estimator_)
print(randm.best_score_)
print(randm.best_params_)
```

```python
randm.score(X_train,Y_train), randm.score(X_test,Y_test)
```

```python
model=CatBoostClassifier(task_type='CPU')
parameters={ 'depth'         : sp_randInt(4,10),
              'learning_rate':sp_randFloat(),
              'iterations'   :sp_randInt(10,100)
           }
randm2=RandomizedSearchCV(estimator=model,param_distributions=parameters, cv=5, n_iter=10, n_jobs=-1)
randm2.fit(X_train,Y_train)
```

```python
randm2.score(X_train,Y_train), randm2.score(X_test,Y_test)
```

```python
parameters={ 'depth'         : sp_randInt(6,10),
              'learning_rate':sp_randFloat(),
              'iterations'   :sp_randInt(10,100),
            # 'colsample_bylevel':[0.3, 0.8, 0.1]
           }
randm3=RandomizedSearchCV(estimator=model,param_distributions=parameters, cv=2, n_iter=10, n_jobs=-1)
randm3.fit(X_train,Y_train)
```

```python
randm3.score(X_train,Y_train), randm3.score(X_test,Y_test)
```

```python
randm3.best_params_
```

```python
from sklearn.model_selection import GridSearchCV

param_grid ={'depth': [randm2.best_params_['depth'],randm2.best_params_['depth']+2,randm2.best_params_['depth']+4],
             'iterations': [randm2.best_params_['iterations'],randm2.best_params_['iterations']+3,randm2.best_params_['iterations']+5],
             'learning_rate': [randm2.best_params_['learning_rate']]
             }
param_grid
```

```python
model = CatBoostClassifier(task_type='CPU')
grid_search=GridSearchCV(estimator=model,param_grid=param_grid,cv=2,n_jobs=-1,verbose=2)
grid_search.fit(X_train,Y_train)
```

```python
grid_search.best_estimator_ 
```

```python
best_grid=grid_search.best_estimator_
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,f1_score
```

```python
y_pred3=best_grid.predict(X_test)
print(confusion_matrix(Y_test,y_pred3))
print("Accuracy Score {}".format(accuracy_score(Y_test,y_pred3)))
print("Classification report: {}".format(classification_report(Y_test,y_pred3)))
```

```python
f1_score(Y_test,y_pred3)
```

```python
best_grid.score(X_train,Y_train), best_grid.score(X_test,Y_test)
```
