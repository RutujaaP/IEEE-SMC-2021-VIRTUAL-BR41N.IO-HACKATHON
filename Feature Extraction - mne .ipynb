```python
import numpy as np
import pandas as pd
from mne_features.feature_extraction import extract_features
import matplotlib.pyplot as plt
```

```python
X_train = np.load("/content/drive/MyDrive/2021_VIIT_P300_Session2 /Hackathon Oct 2021/Epoched Data/X_train.npy")
X_test = np.load("/content/drive/MyDrive/2021_VIIT_P300_Session2 /Hackathon Oct 2021/Epoched Data/X_test.npy")
Y_train = np.load("/content/drive/MyDrive/2021_VIIT_P300_Session2 /Hackathon Oct 2021/Epoched Data/Y_train.npy")
Y_test = np.load("/content/drive/MyDrive/2021_VIIT_P300_Session2 /Hackathon Oct 2021/Epoched Data/Y_test.npy")
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```
